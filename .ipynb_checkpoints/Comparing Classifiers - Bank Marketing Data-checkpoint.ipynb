## Introduction

**In this project, I compare the performance of 4 common classifier models on data from a telemarketing campaign carried out by a Portuguese bank. (The dataset was obtained from the UCI Machine Learning repository: https://archive.ics.uci.edu/ml/datasets/bank+marketing).**

**The telemarketing campaign offered customers incentives to make a deposit to a new account. The classifier model will be designed to use information *available prior to first contact* to predict whether or not a customer ends up making a deposit as a result of the campaign. An effective model of this kind could be used to make future telemarketing campaigns much more effective. This particular campaign, for example, resulted in 4640 deposits out of 41188 customers contacted (a success rate of ~11%), whereas a campaign based on a good classifier model, like the one highlighted in what follows, could be expected to have a success rate (and accompanying profits) 3-4 times as high.**

```python
df = pd.read_csv('bank-additional-full.csv',sep=';')
```

## Implementing Classifier Models

**The initial evaluation metric used for all models below is the F1 score, which factors in both precision and recall. Since the aim is more effective marketing campaigns, it is advisable to seek a good balance between these two values. Increasing the efficiency or success rate of the campaign (i.e. high precision) is helpful, but only if a large enough portion of the customers who would end up making a deposit are reached (i.e. high recall).**

**It is unlikely that maximizing the F1 score captures this balance perfectly, however. The best way to evaluate the models would be to carry out a cost/benefit analysis, using the average cost of successful/unsuccessful contacts and the average benefit of each success to determine the ideal point on the precision-recall curve. A hypothetical version of that kind of evaluation is included below for the best-performing model (as measured by F1 score).**    

```python
# One Hot Encoding and scaling
transformer = make_column_transformer((OneHotEncoder(drop = 'if_binary'), 
                            ['job','marital','education','default','contact','month']),remainder = StandardScaler())
```

```python
X_train, X_test, y_train, y_test = train_test_split(dff.drop('target', axis = 1), dff.target, random_state=22,
                                                   stratify = dff.target)
```

**Model #1: Logistic Regression**

```python
lr_pipe = Pipeline([('transform', transformer), ('lr', LogisticRegression(max_iter=1000,C=0.01))])

params = {'lr__solver': ['lbfgs','sag','newton-cg'],'lr__multi_class': ['auto','ovr','multinomial']}

lr_grid = RandomizedSearchCV(estimator=lr_pipe,param_distributions=params,scoring='f1',cv=5)
lr_grid.fit(X_train,y_train)

print("Best Model: " + str(lr_grid.best_params_))
```

**F1 score & confusion matrix for best model from grid search:**

```python
preds_lr = lr_grid.predict(X_test)
f1_score(y_test,preds_lr)
```

```python
cm = ConfusionMatrixDisplay.from_predictions(y_test,preds_lr,display_labels=['Outcome: No','Outcome: Yes'])
```

**F1 score & confusion matrix for best model after testing different thresholds:**

```python
# get model probabilities for successful outcome:
preds_lr = lr_grid.predict_proba(X_test)[:,1]
```

```python
# test different thresholds for successful outcome prediction:
ts = np.linspace(0.01,0.5,50)
best_f1 = 0.26
best_preds = []
for t in ts:
    preds = np.where(preds_lr>t,1,0)
    f1 = f1_score(y_test,preds)
    if f1 > best_f1:
        best_f1 = f1
        threshold = t
        best_preds = preds
best_preds = best_preds
print('Best threshold: ' + str(threshold))
print('Best F1 score: ' + str(best_f1))
```

```python
confusion_matrix = ConfusionMatrixDisplay.from_predictions(y_test,best_preds,display_labels=['Outcome: No','Outcome: Yes'])
```

**Model #2: k-Nearest Neighbors**

```python
knn_pipe = Pipeline([('transform', transformer), ('knn', KNeighborsClassifier())])

params = {'knn__n_neighbors':[5,13,21,29]}

knn_grid = RandomizedSearchCV(estimator=knn_pipe,param_distributions=params,scoring='f1',cv=5)
knn_grid.fit(X_train,y_train)

print("Best Parameters: " + str(knn_grid.best_params_))
```

**F1 score & confusion matrix for best model from grid search:**

```python
preds_knn = knn_grid.predict(X_test)
f1_score(y_test,preds_knn)
```

```python
confusion_matrix = ConfusionMatrixDisplay.from_predictions(y_test,preds_knn,display_labels=['Outcome: No','Outcome: Yes'])
```

```python
# test different thresholds for successful outcome prediction:
ts = [0.1,0.3]
best_f1 = 0.34
best_preds = []
for t in ts:
    preds = np.where(preds_knn>t,1,0)
    f1 = f1_score(y_test,preds)
    if f1 > best_f1:
        best_f1 = f1
        threshold = t
        best_preds = preds
best_preds = best_preds
print('Best threshold: ' + str(threshold))
print('Best F1 score: ' + str(best_f1))
```

## Hypothetical Cost/Benefit Evaluation

**As noted above, the best way to evaluate these models, from a business perspective, is via cost-benefit analysis. To provide an example of that, I used hypothetical values for the cost of each contact and the net benefit of successful outcomes and then determined the threshold for the Decision Tree model that maximized overall profit. (I used hypothetical values on which the current campaign, not informed by a classifier model, was just barely profitable).**  

```python
from sklearn.metrics import confusion_matrix
```

```python
ts = np.linspace(0.01,0.50,50)
max_profit = 0
profits = []
best_preds = []
# hypothetical cost/benefit values: $11 for unsuccessful contacts (false positives), $100 for successful outcomes (true positives)
cb = np.array([0,-11,0,100])  
for t in ts:
    preds = np.where(preds_dt>t,1,0)
    profit = sum(confusion_matrix(y_test,preds).ravel() * cb)
    profits.append(profit)
    if profit > max_profit:
        max_profit = profit
        best_threshold = t
        best_preds = preds
best_preds = best_preds
max_profit = max_profit
best_threshold = round(best_threshold,2)
print('Best threshold: ' + str(best_threshold))
print('Max profit: ' + str(max_profit))
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=ts,y=profits)
plt.plot(best_threshold,max_profit,'x')
plt.axvline(best_threshold,color='red',alpha=0.4)
plt.text(best_threshold+0.015,max_profit+1000, f'Max Profit of ${max_profit} at Threshold = {best_threshold}',color='red')
plt.title('Profit Curve for Decision Tree Models')
plt.ylim(10000,60000)
plt.xlabel('Threshold Used for Positive Prediction')
plt.ylabel('Total Profit (on Test Data)')
```

**Confusion matrix for the best decision tree model according to the hypothetical cost/benefit analysis:**

```python
cm = ConfusionMatrixDisplay.from_predictions(y_test,best_preds,display_labels=['Outcome: No','Outcome: Yes'])
```

## Findings / Recommendations

**The goal of this project was to construct an effective classifier model based on data from a telemarketing campaign carried out by a Portuguese bank. The performance of 4 common classifier models was examined and compared, with the following results:**

**The best model, using the F1 score as an evaluation metric, was a decision tree of depth 10 and a threshold probability (for positive prediction) of 0.22. Virtually equally good was a multinomial logistic regression model with 'sag' as the solver and a threshold probability (for positive prediction) of 0.20, while the best k-nearest neighbors and support vector machine models were somewhat less effective.** 

**In all cases, prediction was rather difficult, with F1 scores never exceeding 0.5. This is because the models were restricted to information *available prior to first contact* in order to analyze their usefulness at that initial stage. (It was clear in this dataset that a customer *previously accepting* a telemarketing offer is extremely useful information, which should definitely be incorporated into marketing decisions, but information of that kind was not used when constructing/testing these models).** 

**Nonetheless, use of a classifier model with the predictive success obtained here could greatly increase the profit margin of a telemarketing campaign. If we assume, for example, that the campaign actually conducted was only mildly profitable - with profits of roughly 15,000 for a pool of roughly 10,000 customers (the size of the test set used above) - then a more targeted campaign built on the results of the best-performing model could be projected to yield profits 3-4 times as high (roughly 55,000). As such, it is highly recommended that banks make use of classifier models - in this case, a decision tree model as described above - to maximize the effectiveness of their telemarketing campaigns.**
